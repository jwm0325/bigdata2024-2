```python
import pandas as pd
```

```python
경제성장률 = pd.read_excel('도시별경제성장률.xlsx')
경제성장률
```

```python
출산율 = pd.read_excel('도시별출산율.xlsx')
출산율
```

```python
경제성장률.drop(index=0,axis=0,inplace=True)
경제성장률.head()
```

```python
출산율.drop(index=0,axis=0,inplace=True)
출산율.head()
```

```python
경제성장률.rename(columns={"행정구역별(1)":"시군구별"},inplace=True)
경제성장률.head()
```

```python
경제성장률.reset_index(drop=True,inplace=True)
경제성장률.head()
```

```python
출산율.reset_index(drop=True,inplace=True)
출산율.head()
```

```python
전년도대비경제성장률 = 경제성장률[['시군구별']]
전년도대비경제성장률
```

```python
전년도대비경제성장률['2017년'] = 경제성장률['2017'] - 경제성장률['2016']
전년도대비경제성장률['2018년'] = 경제성장률['2018'] - 경제성장률['2017']
전년도대비경제성장률['2019년'] = 경제성장률['2019'] - 경제성장률['2018']
전년도대비경제성장률['2020년'] = 경제성장률['2020'] - 경제성장률['2019']
전년도대비경제성장률['2021년'] = 경제성장률['2021'] - 경제성장률['2020']
전년도대비경제성장률['2022년'] = 경제성장률['2022'] - 경제성장률['2021']
전년도대비경제성장률
```

```python
전년도대비출산율 = 출산율[['시군구별']]
전년도대비출산율
```

```python
전년도대비출산율['2017년'] = 출산율['2017'] - 출산율['2016']
전년도대비출산율['2018년'] = 출산율['2018'] - 출산율['2017']
전년도대비출산율['2019년'] = 출산율['2019'] - 출산율['2018']
전년도대비출산율['2020년'] = 출산율['2020'] - 출산율['2019']
전년도대비출산율['2021년'] = 출산율['2021'] - 출산율['2020']
전년도대비출산율['2022년'] = 출산율['2022'] - 출산율['2021']
전년도대비출산율
```

```python
전년도대비경제성장률['평균'] = 전년도대비경제성장률.drop('시군구별',axis=1).mean(axis='columns')
전년도대비경제성장률
```

```python
전년도대비경제성장률.sort_values('평균').head()
```

```python
전년도대비경제성장률.sort_values('평균',ascending=False).head()
```

```python
최대최소출산율 = 전년도대비출산율.query('시군구별 in ["세종특별자치시","경상남도","울산광역시"]')
최대최소출산율
```

```python
최대최소경제성장률 = 전년도대비경제성장률.query('시군구별 in ["세종특별자치시","경상남도","울산광역시"]').drop('평균',axis=1)
최대최소경제성장률
```

```python
pd.concat([최대최소출산율,최대최소경제성장률])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = {
    "연도": [2017, 2018, 2019, 2020, 2021, 2022],
    "출산율": [-1.53, -1.02, -0.94, -1.95, 0.0, -1.56],
    "경제성장률": [-2.300,-3.100,4.000,-4.500,5.000,-5.2000]
}
```

```python
sns.lineplot(data=df, x='연도', y='출산율', label='출산율', color='blue', marker='o')
sns.lineplot(data=df, x='연도', y='경제성장률', label='경제성장률', color='red', marker='o')

plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('세종의 연도별 출산율과 경제성장률 비교', fontsize=14)
plt.xlabel('연도')
plt.ylabel('출산율 / 경제성장률')
plt.ylim(-10,10)
plt.legend()
plt.grid()
plt.show()
```
